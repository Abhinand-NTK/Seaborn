displot , pariplot, joinplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns = ['sex','smoker','day','time']).corr()
```

```python
sns.heatmap(df.drop(columns = ['sex','smoker','day','time']).corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

categorial plots

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data =df,palette='rainbow')
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
